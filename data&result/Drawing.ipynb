```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style

# matplotlib.rcParams['text.usetex'] = True   # 开启Latex风格
# plt.figure(figsize=(10, 10), dpi=70)    # 设置图像大小
```

```python
theta = np.load('thetas.npy')
losses = np.load('losses.npy')
theta_lm = np.load('thetas_lm.npy')

dtheta = 100*np.diff(theta,axis=0)
dtheta_lm = 100*np.diff(theta_lm,axis=0)
```

# 依次绘制我们算法中每个关节角度数据

```python
colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'navy']
# 绘图
for i in range(theta.shape[1]):
    plt.plot(theta[:, i], label=f'joint {i+1}',color=colors[i])

plt.title('Our Algorithm')
plt.xlabel('Point')
plt.ylabel("Joint Angles (rad)")

# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')   
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=dtheta.shape[1])
plt.grid(linestyle='--', color='black', alpha=0.5)
plt.legend()
plt.show()
```

```python
# 依次绘制每个关节角速度数据
for i in range(dtheta.shape[1]):
    plt.plot(dtheta[:, i], label=f'joint {i+1}',color=colors[i])

plt.title('Our Algorithm')
plt.xlabel('Point')
plt.ylabel("Joint Angular Velocity (rad/s)")
plt.grid(linestyle='--', color='black', alpha=0.5)
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')   
# plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=dtheta.shape[1])

# plt.legend()
plt.show()
```

# 绘制LM算法中的角速度数据

```python
#绘制LM算法中的角速度数据

for i in range(theta_lm.shape[1]):
    plt.plot(theta_lm[:, i], label=f'joint {i+1}',color=colors[i])

plt.title('L-M Algorithm')
plt.xlabel('point')
plt.ylabel("Joint Angle (rad)")
plt.grid(linestyle='--', color='black', alpha=0.5)
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')   
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=dtheta.shape[1])
plt.grid(linestyle='--', color='black', alpha=0.5)
plt.legend()
plt.show()
```

```python
#绘制LM算法中的角速度数据

for i in range(dtheta_lm.shape[1]):
    plt.plot(dtheta_lm[:, i], label=f'joint {i+1}',color=colors[i])

plt.title('L-M Algorithm')
plt.xlabel('point')
plt.ylabel("Joint Angular Velocity (rad/s)")
plt.grid(linestyle='--', color='black', alpha=0.5)
# plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')   
# plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=dtheta.shape[1])

# plt.legend()
plt.show()
```

## 分别对每一个关节进行比较

```python
fig, axs = plt.subplots(12, 1, figsize=(8, 40))

# 循环遍历每一列，将其绘制到对应的子图中
for i in range(12):
    axs[i].plot(theta[:, i], label='Our Algorithm')
    #axs[i].set_xlabel('point')
    axs[i].set_ylabel("Joint Angle (rad)")
    axs[i].grid(linestyle='--', color='black', alpha=0.5)
    axs[i].plot(theta_lm[:, i], label='L-M Algorithm')
    axs[i].set_title('joint {}'.format(i+1))
    axs[i].legend()
```

```python
fig, axs = plt.subplots(12, 1, figsize=(8, 40))

# 循环遍历每一列，将其绘制到对应的子图中
for i in range(12):
    axs[i].plot(dtheta[:, i], label='Our Algorithm')
    axs[i].plot(dtheta_lm[:, i], label='L-M Algorithm')
    #axs[i].set_xlabel('point', labelpad=-10)
    axs[i].set_ylabel("Joint Angular Velocity (rad/s)")
    axs[i].grid(linestyle='--', color='black', alpha=0.5)
    axs[i].set_title('joint {}'.format(i+1))
    axs[i].legend()
```

# 比较均值和极差

```python
# 计算每列数据的最大值减最小值
ranges = dtheta.ptp(axis=0)

# 绘制柱状图
plt.bar(range(1, 13), ranges)

# 添加x轴和y轴标签
plt.xlabel('Joint')
plt.ylabel('Range of angular velocity (rad/s)')

# 显示图形
plt.show()
```

```python

# 计算每列数据的平均值
means = dtheta.mean(axis=0)

# 绘制柱状图
plt.bar(range(1, 13), means)

# 添加x轴和y轴标签
plt.xlabel('Joint')
plt.ylabel('Mean of angular velocity (rad/s)')

# 显示图形
plt.show()
```
